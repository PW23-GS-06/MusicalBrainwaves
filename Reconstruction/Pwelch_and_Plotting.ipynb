## Event e7_s23_c2

```python
import numpy as np
from scipy.signal import welch
import matplotlib.pyplot as plt
import json
```

```python
# Generate random EEG data for demonstration
# np.random.seed(0)
# eeg_data = np.random.randn(5000)  # Replace this with your actual EEG data
```

```python
# Sample EEG data (replace this with your actual EEG data)
file_path="/content/drive/MyDrive/CAPSTONE_626_645_648_651/Data/OpenMIIR/eeg/64_InputChannels/Sub14/P14-formatted.json"
with open(file_path, 'r') as file_x:
  sub_eeg_dict=json.load(file_x)
eeg_data=np.array(sub_eeg_dict["e7_s23_c2"])
```

```python
for key,value in sub_eeg_dict.items():
  print(key)
```

```python
eeg_data=eeg_data.T
```

```python
eeg_data.shape
```

```python
fs = 64 # Sampling frequency in Hz
nperseg = 64  # Window size for each segment
noverlap = 0  # Overlap between segments
```

```python
frequencies, psd = welch(eeg_data,fs=fs, nperseg=nperseg,noverlap= noverlap, scaling='spectrum')
```

```python
print(psd)
```

```python
print(len(psd))
print(len(psd[0]))
```

```python
print(len(frequencies))
```

```python
print(frequencies)
```

```python
for i in range(64):
  # Plot the PSD
  plt.figure(figsize=(10, 5))
  plt.semilogy(frequencies, psd[i])
  plt.title('Power Spectral Density (PSD) of EEG Signal '+str(i))
  plt.xlabel('Frequency (Hz)')
  plt.ylabel('Power/Frequency (dB/Hz)')
  plt.grid()
  plt.show()
```

```python
average_psd = np.mean(psd, axis=0)  # Compute the average PSD across segments

plt.figure(figsize=(10, 6))
plt.semilogy(frequencies, average_psd)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power/Frequency (dB/Hz)')
plt.title('Average Power Spectral Density (PSD) of EEG Data')
plt.grid(True)
plt.show()
```
