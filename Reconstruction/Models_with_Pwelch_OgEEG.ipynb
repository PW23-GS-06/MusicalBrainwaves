```python
import numpy as np
from scipy.signal import welch
import matplotlib.pyplot as plt
import json
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Sample EEG data (replace this with your actual EEG data)
file_path="/content/drive/MyDrive/CAPSTONE_626_645_648_651/Data/OpenMIIR/eeg/64_InputChannels/Sub14/P14.json"
with open(file_path, 'r') as file_x:
  sub_eeg_dict=json.load(file_x)
eeg_data=np.array(sub_eeg_dict["e16_s1_c2"])
eeg_data=eeg_data.T
fs = 64 # Sampling frequency in Hz
nperseg = 64  # Window size for each segment
noverlap = 0  # Overlap between segments
frequencies, psd = welch(eeg_data,fs=fs, nperseg=nperseg,noverlap= noverlap, scaling='spectrum')
for i in range(64):
  # Plot the PSD
  plt.figure(figsize=(10, 5))
  plt.semilogy(frequencies, psd[i])
  plt.title('Power Spectral Density (PSD) of EEG Signal '+str(i))
  plt.xlabel('Frequency (Hz)')
  plt.ylabel('Power/Frequency (dB/Hz)')
  plt.grid()
  plt.show()
```

## For All subjects

```python
PNUMs=['01','04','06','07','09','11','12','13','14']
mega_eeg_list = []
mega_eeg_keys_list = []
mega_audio_psd_list = []
mega_eeg_psd_list = []
```

```python
for PNUM in PNUMs:
  file_path="/content/drive/MyDrive/CAPSTONE_626_645_648_651/Data/OpenMIIR/eeg/64_InputChannels/Sub"+PNUM+"/P"+PNUM+".json"
  with open(file_path, 'r') as file_x:
    sub_eeg_dict=json.load(file_x)
  mega_eeg_list+= [value for value in sub_eeg_dict.values()]
  mega_eeg_keys_list += ["P"+PNUM+"_"+key for key in sub_eeg_dict.keys()]


  audio_path = "/content/drive/MyDrive/CAPSTONE_626_645_648_651/Data/OpenMIIR/audio/PSD/v1_and_v2_psd.json"
  # os.makedirs(prediction_dir, exist_ok=True)
  with open(audio_path, 'r') as file_x:
    env_psd_dict=json.load(file_x)


  for key,value in sub_eeg_dict.items():
    song_id=key.split('_')[1]
    if PNUM in ['01','04','06','07'] and song_id in ['s1','s2','s11','s12','s22']:
      env_key=song_id+'_v1'
    else:
      env_key=song_id+'_v2'
    mega_audio_psd_list.append(env_psd_dict[env_key])
```
