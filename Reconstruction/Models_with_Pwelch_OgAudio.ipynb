## v1_and_v2

```python
import numpy as np
from scipy.signal import welch
import matplotlib.pyplot as plt
import json
```

```python
# Sample EEG data (replace this with your actual EEG data)
file_path="/content/drive/MyDrive/CAPSTONE_626_645_648_651/Data/OpenMIIR/audio/pretrained_vlaai_envelope/v1_and_v2.json"
with open(file_path, 'r') as file_x:
  sub_audio_dict=json.load(file_x)
audio_data=np.array(sub_audio_dict["s1_v2"])
audio_data=audio_data.T
fs = 64 # Sampling frequency in Hz
nperseg = 64  # Window size for each segment
noverlap = 0  # Overlap between segments
frequencies, psd = welch(audio_data,fs=fs, nperseg=nperseg,noverlap= noverlap, scaling='spectrum')
# for i in range(64):
  # Plot the PSD
plt.figure(figsize=(10, 5))
plt.semilogy(frequencies, psd)
plt.title('Power Spectral Density (PSD) of Audio Signal')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power/Frequency (dB/Hz)')
plt.grid()
plt.show()
```

```python
sub_psd_audio_dict={}
for key,value in sub_audio_dict.items():
  audio_data=np.array(sub_audio_dict[key])
  audio_data=audio_data.T
  fs = 64 # Sampling frequency in Hz
  nperseg = 64  # Window size for each segment
  noverlap = 0  # Overlap between segments
  frequencies, psd = welch(audio_data,fs=fs, nperseg=nperseg,noverlap= noverlap, scaling='spectrum')
  sub_psd_audio_dict[key]=psd.tolist()
```

```python
output_file_path="/content/drive/MyDrive/CAPSTONE_626_645_648_651/Data/OpenMIIR/audio/PSD/v1_and_v2_psd.json"
with open(output_file_path, 'w') as file_x:
  json.dump(sub_psd_audio_dict, file_x)
```
